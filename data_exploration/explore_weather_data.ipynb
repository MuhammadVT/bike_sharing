# Explore _weather_ Data

In this short report we explore the _weather_ data.

```python
# load necessary packages
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# show the figures within the notebook
%matplotlib inline

# Select ggplot as style
plt.style.use("ggplot")
```

## Display the data

```python
# read the data
df_weather = pd.read_csv("../data/weather.csv", parse_dates=["date"])

# display the data
display(df_weather.head())

# show the number of data points in trip.csv
print("This table has {npnts} data points and {col} columns.".format(npnts=df_weather.shape[0], col=df_weather.shape[1]))

# display the features
print("\nThe columns in this dataset are:")
print("".join([str(i+1) + ". " + df_weather.columns[i]+"\n" for i in range(len(df_weather.columns))]))

# types of events
print("\nThe types of weather events in the 'events' column are:")
evnts = [str(x) for x in df_weather.events.unique()]
print("".join([str(i+1) + ". " + evnts[i] + "\n" for i in range(len(evnts))]))
```

```python
# Produce a scatter matrix for each pair of features in the data
dfn = df_weather.iloc[:, [1,2,3]]
pd.plotting.scatter_matrix(dfn, alpha = 0.5, figsize = (14,8), diagonal = 'kde');
```

```python
# create a figure
fig, ax = plt.subplots()

ax.matshow(df_weather.corr())
```
